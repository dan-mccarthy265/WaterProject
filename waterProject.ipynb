```python
import csv
import torch
import torch.nn as nn
import torch.nn.functional as F
from torch.optim import SGD
from torch.nn import MSELoss
from torch.nn import Sigmoid
from torch.nn import Linear
import numpy  as np

f = "drinking_water_potability.csv"
training = []
validation = []
test = []
fields = []
with open (f, 'r') as csvfile:
    csvreader = csv.reader(csvfile)
    fields = next(csvreader)
    count = 0
    for row in csvreader:
        fulldata = 1
        for i in range(len(row)):
            if row[i] == '':
                fulldata = 0 
            else:
                row[i] = float(row[i])
        if fulldata == 1:
            if count < 6:
                training.append(row)
            elif count < 8:
                validation.append(row)
            elif count < 10:
                test.append(row)
            count += 1
            if count == 10:
                count = 0
            

#print(training)


class Model(nn.Module):
    # define model elements
    def __init__(self):
        super(Model, self).__init__()
        self.layer = Linear(9, 18)
        self.hidden = Linear(18, 18)
        self.hidden2 = Linear(18, 9)
        self.hidden3 = Linear(9, 1)
        
        
        self.activation = Sigmoid()
 
    # forward propagate input
    def forward(self, X):
        X = self.layer(X)
        X = self.activation(X)
        X = self.hidden(X)
        X = self.activation(X)
        X = self.hidden2(X)
        X = self.activation(X)
        X = self.hidden3(X)
        return X

model = Model()

criterion = MSELoss()
optimizer = SGD(model.parameters(), lr=0.001, momentum=0.9)


for epoch in range(1000):
    for i in range(len(training)):
        optimizer.zero_grad()
        inputs = training[i][:-1]
        inputs = torch.tensor(inputs)
        target = training[i][9]
        target = torch.tensor(target)
        yhat = model(inputs)
        loss = criterion(yhat, target)
        loss.backward()
        optimizer.step()
    if (epoch%100 == 0):
        print('loss at epoch ' + str(epoch) + ' = ' + str(float(loss)))
print('Final loss: ' + str(float(loss)))


val = torch.FloatTensor(validation)

for i in range(len(val)):
    inputs = val[i][:-1]
    inputs = torch.tensor(inputs)
    pred = model(inputs)
    print('inputs: ' + str(inputs) + '\n' + 'Prediction: ' + str(float(pred[0])) + '\n' + 'Actual: ' + str(val[i][9]))
        
print(fields)
            
```
